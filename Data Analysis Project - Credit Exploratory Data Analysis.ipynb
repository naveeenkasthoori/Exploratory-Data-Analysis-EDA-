### - The bank could market more loans to people with an academic degree since as per our analysis, they seem to have very few difficulties with payments. 

### - The bank could market more loans to people who are either businessmen or students. They too seem to encounter very few hassles in timely payments.

### - The bank could process more loans if the purpose is 'buying a garage' or 'Journey' since bvery few people in these segments have payment difficulties. 

### - The bank could market more 'education loans as well as there is a strong indication that people who apply for such loans have only few, if any,  payment diffculties. 

### - Coupling the above fact with the fact that people with academic degrees tend to have less payment diffuclties, education loans for the purpose of obtaining an academic degree seems to be a segment they should be bullish about.
